```python
import pandas as pd


home_data = pd.read_csv('Metabolic Syndrome.csv',usecols = ['HDL', 'Triglycerides', 'MetabolicSyndrome'])
home_data.reset_index()
home_data.head()
```

```python
import seaborn as sns

sns.scatterplot(data = home_data, x = 'HDL', y = 'Triglycerides', hue = 'MetabolicSyndrome')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(home_data[['HDL', 'Triglycerides']], home_data[['MetabolicSyndrome']], test_size=0.33, random_state=0)
```

```python
from sklearn import preprocessing

X_train_norm = preprocessing.normalize(X_train)
X_test_norm = preprocessing.normalize(X_test)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(X_train_norm)
```

```python
sns.scatterplot(data = X_train, x = 'HDL', y = 'Triglycerides', hue = kmeans.labels_)
```
